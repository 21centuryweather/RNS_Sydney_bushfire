```python
import geopandas as gpd
import matplotlib.pyplot as plt
import xarray as xr
```

```python
burn_area = gpd.read_file("/g/data/ng72/as9583/control/merged_burn_area.gpkg")

datapath = f'/g/data/fy29/mjl561/cylc-run/rns_ostia_2019_bushfire/netcdf'
variable = 'surface_altitude'
sa1 = xr.open_dataset(f'{datapath}/{variable}/E5L_1_CCI_WC_{variable}.nc')[variable].isel(time=0)
```

```python
fig, ax = plt.subplots(figsize=(10, 8))

burn_area.plot(ax=ax, edgecolor='black', facecolor='orange', alpha=0.5)

contours2 = sa1.plot.contour(
    ax=ax,
    levels=range(0, 2000, 1000),
    colors='black',
    linewidths=1,
    add_colorbar=False
)
plt.title("Fires scars in 1km model domain")
plt.xlabel("Longitude ($\\degree$E)", fontsize=12)
plt.ylabel("Latitude ($\\degree$N)", fontsize=12)

plt.show()
```
